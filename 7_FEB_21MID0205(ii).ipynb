```python
from urllib import request
url = "http://www.gutenberg.org/files/2554/2554-0.txt"
response = request.urlopen(url)
raw = response.read().decode('utf8')
```

```python
print(type(raw))
print(len(raw))
raw[:75]
```

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(raw)
print(type(tokens))
print(len(tokens))
```
